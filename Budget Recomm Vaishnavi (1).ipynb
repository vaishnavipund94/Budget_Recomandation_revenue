```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(x=df["REVENUE_GROUP_TYPE"], data=df,color ="blue")
plt.title("Count Plot for REVENUE_GROUP_TYPE type")
plt.xticks(rotation = 85)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='REVENUE_GROUP_TYPE', y='ESTIMATED_REVENUE', data=df)
plt.xticks(rotation=45)
plt.title("Revenue Distribution by Group Type")
plt.show()
```

- **Intergovernmental Revenue has the highest estimated revenue.**
- **Proceeds and Transfers second highest estimated revenue.**
- **Local Non-Tax Revenue has a moderate estimated revenue level.**
- **Local Tax Revenue has the lowest estimated revenue.**

```python
group_data = df.groupby('REVENUE_GROUP_TYPE')['ESTIMATED_REVENUE'].sum()

plt.figure(figsize=(8,8))
plt.pie(group_data, labels=group_data.index, autopct='%1.1f%%', startangle=90)
plt.title("Revenue Share by Group Type")
plt.show()
```

- **Local Tax Revenue has the highest share at 78.3%.**
- **Intergovernmental Revenue contributes 5.3%.**
- **Proceeds and Transfers account for 4.8%.**
- **Local Non-Tax Revenue has the lowest share at 11.6%.**

```python
sns.regplot(x=df.index, y="ESTIMATED_REVENUE", data=df)
plt.title("Record Index vs Estimated Revenue")
plt.xlabel("Index")
plt.ylabel("Estimated Revenue")
plt.show()
```

- **The Visualization here indicates an downward trend in estimated revenue aross index records.**
- **which means that there is a negative relationship as the record index decreases the estimated revenue also decreases.**
