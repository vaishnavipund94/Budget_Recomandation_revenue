```python
import matplotlib.pyplot as plt
```

```python
plt.bar(df['REVENUE_GROUP_TYPE'],df['ESTIMATED_REVENUE'], color ="skyblue")
plt.title("Estimated revenue by revenue group type")
plt.xlabel('REVENUE_GROUP_TYPE')
plt.ylabel('ESTIMATED_REVENUE')
plt.xticks(rotation=65)
plt.show()
```

- **Local tax** has the **highest estimated revenue** in revenue group above 2.00**
- **Local Non-Tax** has the **lowest estimated revenue below 0.50**
- **Intergovernmental revenue** and **Proceeds and Transfers In** has almost similar revenue **above 0.50**

```python
plt.bar(df['REVENUE_CATEGORY'],df['ESTIMATED_REVENUE'], color ="skyblue")
plt.title("Estimated revenue by revenue category")
plt.xlabel('REVENUE_CATEGORY')
plt.ylabel('ESTIMATED_REVENUE')
plt.xticks(rotation=90)
plt.show()
```

- Recereation taxes has the highest estimated revenue in around 2.00
- **some categories have less revenue**
- leases rentals and sales has the **very lowest estimated revenue in around 0.10.**

```python
plt.figure(figsize=(12,6))
plt.bar(df['FUND_NAME'], df['ESTIMATED_REVENUE'])
plt.xticks(rotation=75, fontsize=10)
plt.xlabel("Fund Name")
plt.ylabel("Estimated Revenue")
plt.title("Estimated Revenue by Fund")
plt.show()
```

```python
group_data = df.groupby('REVENUE_GROUP_TYPE')['ESTIMATED_REVENUE'].sum()

plt.figure(figsize=(8,5))
plt.bar(group_data.index, group_data.values)
plt.xlabel("Revenue Group Type")
plt.ylabel("Total Estimated Revenue")
plt.title("Total Revenue by Group Type")
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(8,5))
sns.boxplot(x='REVENUE_GROUP_TYPE', y='ESTIMATED_REVENUE', data=df)
plt.xticks(rotation=45)
plt.title("Revenue Distribution by Group Type")
plt.show()
```

- Intergovernmental Revenue has the highest estimated revenue.
- Proceeds and Transfers second highest estimated revenue.
- Local Non-Tax Revenue has a moderate estimated revenue level.
- Local Tax Revenue has the lowest estimated revenue.

```python
group_data = df.groupby('REVENUE_GROUP_TYPE')['ESTIMATED_REVENUE'].sum()

plt.figure(figsize=(8,8))
plt.pie(group_data, labels=group_data.index, autopct='%1.1f%%', startangle=90)
plt.title("Revenue Share by Group Type")
plt.show()
```

- **Local Non-Tax Revenue has the highest share at 78.3%.**
- **Intergovernmental Revenue contributes 11.6%.**
- **Proceeds and Transfers account for 5.3%.**
- **Local Tax Revenue has the lowest share at 4.8%.**
